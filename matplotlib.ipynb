```python
import pandas as pd
import matplotlib.pyplot as plt
dicts = {"countries":['india','uk','canada','USA','japan'],"valume":[2,1,1.5,1.9,2],"Amount":[3,2.5,5,6,1],"papulation":[1.3,0.5,1,2,2.5]}
card = pd.DataFrame(dicts)
plt.bar(card['countries'],width=0.1,height=0.09,color='red')
```

```python
plt.pie(card['valume'])
```

```python
plt.pie(card['valume'],labels=card['countries'])
```

```python
plt.pie(card['valume'],labels=card['countries'],autopct='%1.2f%%')
```

```python
plt.pie(card['valume'],labels=card['countries'],autopct='%1.2f%%',explode=(0,0,0.1,0,0.2))
```

```python
plt.pie(card['valume'],labels=card['countries'],autopct='%1.2f%%',explode=(0,0,0.1,0,0.2),shadow=True)
```

```python
plt.pie(card['valume'],labels=card['countries'],autopct='%1.2f%%',explode=(0,0,0.1,0,0.2),shadow=True,colors=['red','blue','green','yellow','orange'])
```

```python
plt.pie(card['valume'],labels=card['countries'],autopct='%1.2f%%',explode=(0,0,0.1,0,0.2),shadow=True,colors=['red','blue','green','yellow','orange'])
plt.title('number of customer buy countries')
```

```python
# Sactter plot
import pandas as pd
import matplotlib.pyplot as plt
dicts = {"countries":['india','uk','canada','USA','japan'],"valume":[2,1,1.5,1.9,2],"Amount":[3,2.5,5,6,1],"papulation":[1.3,0.5,1,2,2.5]}
card = pd.DataFrame(dicts)
plt.scatter(card['countries'],card['Amount'])
```

```python
plt.scatter(card['countries'],card['Amount'],s=200)
```

```python
plt.scatter(card['countries'],card['Amount'],s=200,color='red')
```

```python
plt.scatter(card['countries'],card['Amount'],s=200,color='red',marker='^')
```

```python
plt.scatter(card['countries'],card['papulation'],s=200,color='red',marker='^')
plt.legend(['Amount','countries','papulation','valume'])
```

```python
plt.scatter(card['countries'],card['papulation'],s=200,color='red',marker='^')
plt.xlabel('countries')
plt.ylabel('papulation')
```

```python
plt.scatter(card['countries'],card['papulation'],s=200,color='red',marker='^')
plt.scatter(card['countries'],card['Amount'],s=200,color='blue',marker='>')

plt.xlabel('countries')
plt.ylabel('papulation')
plt.title("cards of info")
```

```python
# Box plot/whisker plot

import pandas as pd
import matplotlib.pyplot as plt
dicts = {"countries":['india','uk','canada','USA','japan'],"valume":[2,1,1.5,1.9,2],"Amount":[3,2.5,5,6,1],"papulation":[1.3,0.5,1,2,2.5]}
card = pd.DataFrame(dicts)
plt.boxplot(card['Amount'])
```

```python
plt.boxplot(card['Amount'],sym='b+')
```

```python
plt.boxplot(card['Amount'],sym=' ')
```

```python
plt.boxplot(card['Amount'],sym='b+',showmeans=True)
```

```python
plt.boxplot(card['Amount'],sym='b+',showmeans=True,meanline=True)
```

```python
plt.boxplot(card['Amount'],sym='b+',showmeans=True,meanline=True,vert=False)
```

```python
plt.boxplot(card['Amount'],sym='b+',showmeans=True,meanline=True,vert=False)
plt.boxplot(card['valume'],sym='',showmeans=True,meanline=True,vert=True)
```

```python
plt.boxplot(card[['valume','Amount','papulation']],sym='',showmeans=True,meanline=True,vert=True,labels=['Amount','papulation','valume'])
```

```python
#line graphs

import pandas as pd
import matplotlib.pyplot as plt
dicts = {"countries":['india','uk','canada','USA','japan'],"valume":[2,1,1.5,1.9,2],"Amount":[3,2.5,5,6,1],"papulation":[1.3,0.5,1,2,2.5]}
card = pd.DataFrame(dicts)
plt.plot(card["countries"],card['valume'])
```

```python
# multiple lines
plt.plot(card["countries"],card[['valume','papulation','Amount']])
```

```python
#legends
plt.plot(card["countries"],card[['valume','papulation','Amount']])
plt.legend(['valume','papulation','Amount'])
```

```python
#marker
plt.plot(card["countries"],card[['valume','papulation','Amount']],marker='o')
```

```python
#marker size
plt.plot(card["countries"],card[['valume','papulation','Amount']],marker='o',markersize=10)
```

```python
# linewidth
plt.plot(card["countries"],card[['valume','papulation','Amount']],marker='o',markersize=10,linewidth=5)
```

```python
#marker face color
plt.plot(card["countries"],card[['valume','papulation','Amount']],marker='o',markersize=10,linewidth=5,markerfacecolor='black')
```

```python
#marker edgecolor
plt.plot(card["countries"],card[['valume','papulation','Amount']],marker='o',markersize=10,linewidth=5,markerfacecolor='white',markeredgecolor='black')
```

```python
#linestyle
# Axis labels
plt.plot(card["countries"],card[['valume','papulation','Amount']],marker='o',markersize=10,linewidth=5,markerfacecolor='white')
plt.xlabel("countries")
plt.ylabel("Amount")

```

```python
# grid
plt.plot(card["countries"],card[['valume','papulation','Amount']],marker='o',markersize=10,linewidth=5,markerfacecolor='white')
plt.grid(True)
plt.title('cards information')
```

```python
# stack plot or Area chart
import pandas as pd
import matplotlib.pyplot as plt
dict1 = {"year":[2015,2016,2017,2018],"cards":[2,1.5,6,3],"VAF":[1.3,0.03,0.5,0.27]}
prod = pd.DataFrame(dict1)
plt.stackplot(prod["year"],prod['cards'])
```

```python
#multiple
plt.stackplot(prod["year"],prod['cards'],prod['VAF'])
```

```python
#legends
plt.stackplot(prod["year"],prod['cards'],prod['VAF'])
plt.legend(['year','cards','VAF'])
```

```python
# colors
plt.stackplot(prod["year"],prod['cards'],prod['VAF'],colors=['red','green'])
plt.legend(['year','cards','VAF'])
```
